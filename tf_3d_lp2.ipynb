```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
with tf.device("/gpu:1"):
    weights=np.load("weights.npy",allow_pickle=True)
    Oadam = tf.keras.optimizers.Adam(
        learning_rate=0.001,
        name='Adam'
    )

    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Dense(64, input_shape=(3,), activation="relu"))
    model.add(tf.keras.layers.Dense(64, activation="relu"))
    model.add(tf.keras.layers.Dense(128, activation="relu"))
    model.add(tf.keras.layers.Dense(512, activation="relu"))
    model.add(tf.keras.layers.Dense(256, activation="relu"))
    model.add(tf.keras.layers.Dense(1))

    model.summary()

    model.compile(optimizer=Oadam, loss='mse')

    model.set_weights(weights)
```

```python
with tf.device("/gpu:1"):
    SIZE = [[0, 0, 0], [1168, 1168, 180]]

    value_np = model(tf.convert_to_tensor([[x, y, SIZE[0][2]] for y in range(SIZE[0][1], SIZE[1][1] + 1) for x in range(SIZE[0][0], SIZE[1][0] + 1)])).numpy()
    for z in range(SIZE[0][2] + 1, SIZE[1][2] + 1):
        print(z)
        space_tensor = tf.convert_to_tensor([[x, y, z] for y in range(SIZE[0][1], SIZE[1][1] + 1) for x in range(SIZE[0][0], SIZE[1][0] + 1)])
        value_np = np.concatenate((value_np, model(space_tensor).numpy()), axis=0)
```

```python
value_np = value_np.reshape((SIZE[1][2] - SIZE[0][2] + 1, SIZE[1][1] - SIZE[0][1] + 1, SIZE[1][0] - SIZE[0][0] + 1))
print(value_np.shape)
np.save("value_lp_fr1.npy", value_np)
```
